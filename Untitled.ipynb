```python
from random import randint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
def weapon (sides, exploding = 0):
    damage = 0
    last_roll=sides
    
    if exploding == 1:
        while last_roll == sides:
            last_roll = randint(1,sides)
            damage += last_roll
    else:
        damage += randint(1,sides)

    return(damage)

def multi_weapon(number, sides, exploding = 0, modifier = 0):
    total_dmg=0
    i=1
    while i <= number:
        total_dmg += weapon(sides,exploding)
        i+=1
    return(total_dmg+modifier)

def create_weapon_dataframe(number,sides,exploding = 0,modifier = 0,iterations=1000):

    dmg_list = []
    i = 0

    while i < iterations:
        dmg_list.append(multi_weapon(number,sides,exploding,modifier))
        i+=1
    
    if ((modifier == 0) & (exploding == 1)):
        dice_name = '{0}D{1} Exploding'.format(number,sides)
    elif ((modifier == 1) & (exploding == 1)):
        dice_name = '{0}D{1} +{2} Exploding'.format(number,sides,modifier)
    elif ((modifier == 1) & (exploding == 0)):
        dice_name = '{0}D{1} +{2}'.format(number,sides,modifier)
    elif ((modifier == 0) & (exploding == 0)):
        dice_name = '{0}D{1}'.format(number,sides)
    
    temp_df  = pd.DataFrame(
        {'Roll' : list(range(1,iterations+1)),
         'Damage' : dmg_list,
         'Dice' : dice_name
        }
    )

    return(temp_df)

def create_table(df,high_dmg):
    table_df = df.copy()
    
    table_df['Over {}'.format(high_dmg)] = (table_df['Damage']>=high_dmg) * 1
    table_df = table_df.groupby(['Dice']).agg({'Damage':['sum','max'] , 'Roll':'count', 'Over {}'.format(high_dmg):['sum','mean']}).reset_index()

    table_df.columns = ['Dice','Total Damage','Max Damage','Rolls','# Over {}'.format(high_dmg),'% Over {}'.format(high_dmg)]
    table_df['Average Damage'] = table_df['Total Damage']/table_df['Rolls']

    display(table_df.loc[:,['Dice','Average Damage','Max Damage','# Over {}'.format(high_dmg),'% Over {}'.format(high_dmg)]])

def create_cm_plt(df,dice_name,modifier):
    plot_df = df.copy()

    fig, axes = plt.subplots(1,1,figsize=(20,10))

    sns.kdeplot(plot_df.loc[plot_df['Dice'] == '{0} Exploding'.format(dice_name),'Damage'],
                cumulative=True,
                color='red',label = '{} Exploding'.format(dice_name))
    sns.kdeplot(plot_df.loc[plot_df['Dice']== '{0} +{1}'.format(dice_name,modifier),'Damage'],
                cumulative=True,
                color = 'blue',label = '{0} +{1}'.format(dice_name,modifier))
    sns.kdeplot(plot_df.loc[plot_df['Dice']== dice_name,'Damage'],
                cumulative=True,
                color = 'green',label = dice_name)
    sns.kdeplot(plot_df.loc[plot_df['Dice']== '{0} +{1} Exploding'.format(dice_name,modifier),'Damage'],
                cumulative=True,
                color = 'orange',label = '{0} +{1} Exploding'.format(dice_name,modifier))
    #plt.axhline(y=0.25)
    plt.yticks(np.arange(0, 1.1, step=0.1))
    plt.grid(True)
    plt.show()
    
def create_box_plot(df):
    bx_plot_df = df.copy()
    
    bx_plot_df = bx_plot_df.loc[:,['Dice','Damage']]
    
    bx_plot_df['Dice'] = pd.Categorical(bx_plot_df['Dice'])
    bx_plot_df['Damage'] = bx_plot_df['Damage'].astype(float)
    
    fig, axes = plt.subplots(1,1,figsize=(20,10))
    sns.boxplot(x='Dice', y='Damage', data=bx_plot_df)
    plt.show()

def dice_comparison (n,s,m,iterations=1000):
    dice_sides = s
    dice_number = n
    mod = m

    weapon_df =  pd.DataFrame(columns = ['Roll','Damage','Dice'])
    weapon_df = weapon_df.append(create_weapon_dataframe(number = dice_number,
                                                         sides = dice_sides,
                                                         exploding = 0,
                                                         modifier = 0,
                                                         iterations=iterations
                                                        ))
    weapon_df = weapon_df.append(create_weapon_dataframe(number = dice_number,
                                                         sides = dice_sides,
                                                         exploding = 0,
                                                         modifier = m,
                                                         iterations=iterations
                                                        ))
    weapon_df = weapon_df.append(create_weapon_dataframe(number = dice_number,
                                                         sides = dice_sides,
                                                         exploding = 1,
                                                         modifier = 0,
                                                         iterations=iterations
                                                        ))
    weapon_df = weapon_df.append(create_weapon_dataframe(number = dice_number,
                                                         sides = dice_sides,
                                                         exploding = 1,
                                                         modifier = m,
                                                         iterations=iterations
                                                        ))


    #Create tables
    high_dmg = ((dice_sides-1)*dice_number)
    create_table(weapon_df,high_dmg)

    #Create cum_df plots
    dice_name = str(dice_number)+'D'+str(dice_sides)
    create_cm_plt(weapon_df,dice_name,m)

    #Create boxplots
    create_box_plot(weapon_df)
```

```python
dice_comparison(2,6,1,100)
```
